```python
# Average Rating per Movie

movie_ratings = ratings.groupby("movieId").agg(
    avg_rating=("rating", "mean"),
    rating_count=("rating", "count")
).reset_index()

movie_ratings.head()
```

```python
# Number of Tags per Movie

movie_tags = tags.groupby("movieId").agg(
    tag_count=("tag", "count")
).reset_index()

movie_tags.head()
```

```python
# Merge Data

master_df = movies.merge(movie_ratings,on="movieId",how="left")
master_df = master_df.merge(movie_tags,on="movieId",how="left")
master_df = master_df.merge(links,on="movieId",how="left")

master_df.head()
```
